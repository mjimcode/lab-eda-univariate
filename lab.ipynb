```python
import pandas as pd
import matplotlib.pyplot as plt

# Carga el dataset
df = pd.read_csv('C:\\Users\\Mario\\Documents\\GitHub\\lab-eda-univariate\\amz_uk_price_prediction_dataset.csv')

# Tabla de frecuencia para la columna 'category'
category_freq = df['category'].value_counts()

# Muestra las 5 categorías más comunes
top_5_categories = category_freq.head(5)
print(top_5_categories)
```

```python
# Diagrama de barras para todas las categorías
category_freq.plot(kind='bar', figsize=(10,6))
plt.title('Distribución de productos por categorías')
plt.xlabel('Categorías')
plt.ylabel('Frecuencia')
plt.show()

# Diagrama de pie para las 5 categorías principales
top_5_categories.plot(kind='pie', autopct='%1.1f%%', figsize=(8,8))
plt.title('Proporción de las 5 categorías principales')
plt.ylabel('')
plt.show()
```

```python
# Medidas de tendencia central para 'price'
mean_price = df['price'].mean()
median_price = df['price'].median()
mode_price = df['price'].mode()[0]

print(f"Media del precio: {mean_price}")
print(f"Mediana del precio: {median_price}")
print(f"Moda del precio: {mode_price}")
```

```python
# Medidas de dispersión para 'price'
var_price = df['price'].var()
std_price = df['price'].std()
range_price = df['price'].max() - df['price'].min()
iqr_price = df['price'].quantile(0.75) - df['price'].quantile(0.25)

print(f"Varianza del precio: {var_price}")
print(f"Desviación estándar del precio: {std_price}")
print(f"Rango del precio: {range_price}")
print(f"Rango intercuartílico (IQR) del precio: {iqr_price}")
```

```python
# Histograma de los precios
plt.hist(df['price'], bins=50, edgecolor='black', alpha=0.7)
plt.title('Distribución de precios')
plt.xlabel('Precio')
plt.ylabel('Frecuencia')
plt.show()

# Boxplot para los precios
plt.boxplot(df['price'])
plt.title('Diagrama de caja para los precios')
plt.ylabel('Precio')
plt.show()
```

```python
# Medidas de tendencia central para 'rating'
mean_rating = df['rating'].mean()
median_rating = df['rating'].median()
mode_rating = df['rating'].mode()[0]

print(f"Media de valoraciones: {mean_rating}")
print(f"Mediana de valoraciones: {median_rating}")
print(f"Moda de valoraciones: {mode_rating}")
```

```python
# Medidas de dispersión para 'rating'
var_rating = df['rating'].var()
std_rating = df['rating'].std()
iqr_rating = df['rating'].quantile(0.75) - df['rating'].quantile(0.25)

print(f"Varianza de valoraciones: {var_rating}")
print(f"Desviación estándar de valoraciones: {std_rating}")
print(f"Rango intercuartílico (IQR) de valoraciones: {iqr_rating}")
```

```python
# Calcula skewness y kurtosis
skewness_rating = df['rating'].skew()
kurtosis_rating = df['rating'].kurt()

print(f"Skewness de valoraciones: {skewness_rating}")
print(f"Kurtosis de valoraciones: {kurtosis_rating}")
```

```python
# Histograma para las valoraciones
plt.hist(df['rating'], bins=10, edgecolor='black', alpha=0.7)
plt.title('Distribución de valoraciones')
plt.xlabel('Valoración')
plt.ylabel('Frecuencia')
plt.show()
```
